# Compute PCA
Note: since the signs of the eigenvectors are essentially arbitrary (e.g. Harman, H.H. Modern Factor Analysis. 3rd Edition, The University of Chicago Press, Chicago, 1976), for a given % of retained variance the PC scores are only saved once to ease the comparison of the results between the lockdown and recovery phase

```python
## PCA
df = df_drop.drop_duplicates('geoid')
X, var_exp = run_ppca(df, ags_vars, ncomponents = len(ags_vars), min_obs = 0.95)
```

## Retain up to 90% of the variance

```python
pc_var_thr = 0.9
```

```python
## Plot explained variance
plot_pc_var(var_exp, textsize = 15, title = 'PPCA explained variance ratio', cum_var_thr = pc_var_thr)
plt.savefig('../plots/{}_{}_var_exp.pdf'.format(filename,pc_var_thr)) 
```

```python
## Plot PC correlations
plot_pc_corr(X, var_exp, ags_vars, var_exp_thr = pc_var_thr, nrows = 20, figsize_w = 60, figsize_h = 60)
plt.savefig('../plots/{}_{}_pc_corr.pdf'.format(filename,pc_var_thr)) 
```

### Save files

```python
## Select only up the PC up to expalain pc_var_thr % of the variance
cum_var_exp = np.cumsum(var_exp)
ncomponents = np.where(cum_var_exp > pc_var_thr)[0][0]
df_drop = X[[i for i in df_drop.columns.to_list() if 'pc_' not in i] + ['pc_' +  str(j) for j in range(ncomponents)]]
```

```python
df_up = df_up.merge(df_drop[['geoid'] + [i for i in df_drop if 'pc_' in i]], on = 'geoid')
```

```python
## Add numeric ID to each county
df_drop['ID'] = pd.factorize(df_drop.geoid)[0]+1
df_up['ID'] = pd.factorize(df_up.geoid)[0]+1
```

```python
df_drop[df_drop.columns[df_drop.columns!='geom']].to_csv("../data/{}_{}.csv".format(filename_drop,pc_var_thr),index = False)
df_up[df_up.columns[df_up.columns!='geom']].to_csv("../data/{}_{}.csv".format(filename_up,pc_var_thr),index = False)
```

## Retain up to 55 % of the variance

```python
pc_var_thr = 0.55
```

```python
## Plot PC correlations
plot_pc_corr(X, var_exp, ags_vars, var_exp_thr = pc_var_thr)
plt.savefig('../plots/{}_{}_pc_corr.pdf'.format(filename,pc_var_thr)) 
```

### Save files

```python
df_up = df_up.drop([i for i in df_up if 'pc_' in i],axis=1).merge(df_drop[['geoid'] + [i for i in df_drop if 'pc_' in i]], on = 'geoid')
```
